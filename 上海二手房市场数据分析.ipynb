```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('houses.csv',encoding='gbk').drop_duplicates()
df.head()
```

1请统计出总房源数量、总小区数量、总版块数量（任务提示：直接统计数据的条数，即用 len 函数即可。阅读参考文章③）

```python
print('总房源数：',len(df.index))
print('总小区数：',len(df.xiaoqu_name.drop_duplicates()))
print('总版块数：',len(df.sub_district_name.drop_duplicates()))
```

2请统计出有哪些不同的户型、有哪些不同的装修情况、有哪些不同的建造年份（任务提示：可用 unique 函数进行统计。阅读参考文章④）

```python
print('户型：',df.huxing.unique())
print('装修：',df.zhuangxiu.unique())
print('建造年份：',df.buildyear.unique())
```

3请计算出上海二手房每平方米均价

```python
df.up_price.mean()
```

4请统计出不同户型对应的房源数量、平均面积。按不同户型的房源数量，从大到小进行排序，判断出主流户型是什么？平均面积多大？（任务提示：以 huxing 为分组对象，展示 size 列，统计户型数量和 size 平均数。阅读参考文章⑤⑥⑦⑧）

```python
df.pivot_table(index='huxing',values='size',aggfunc=[len,np.mean]).sort_values(by=('len','size'),ascending=False)
```

5获取每个小区的房源数量，并按从大到小的顺序排序

```python
df.xiaoqu_name.value_counts()
```

6请统计出上海每个小区每平方米的均价（阅读参考文章⑤⑥⑦⑧）

```python
df.pivot_table(index='xiaoqu_name',values='up_price')
```

7请统计出上海二手房总价排在前 5 位的房源，看看超级豪宅有多壕？

```python
df.sort_values(by='price',ascending=False).head()
```

8请计算出不同年代的房子数量（10年一个区间）（任务提示：先将 buildyear 列转为时间序列，再利用 groupby 函数统计出每年的房源数量，最后用 resample 函数进行汇总。阅读参考文章⑨⑩⑪）

```python
date=pd.to_datetime(df.buildyear,format='%Y')
df.buildyear=date
df.groupby('buildyear').count().resample('10AS').sum()['_id']
```
